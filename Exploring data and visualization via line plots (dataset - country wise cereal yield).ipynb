## Visualizing Data

The primary plotting library we will explore in the course is Matplotlib.
One of the core aspects of Matplotlib is matplotlib.pyplot. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc. 

Let's start by importing Matplotlib and Matplotlib.pyplot as follows:

```python
# we are using the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

check if Matplotlib is loaded.

```python
print ('Matplotlib version: ', mpl.__version__)
```

applying a style to Matplotlib.

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

pandas has a built-in implementation of Matplotlib that we can use. Plotting in pandas is as simple as appending a .plot() method to a series or dataframe.

We can use Line Plots here. A line chart or line plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. We can use line plot when we have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

First, we will extract the data series for a country, for instance, Pakistan.

```python
pakistan = df_agri.loc['Pakistan', years] # passing in years 1961 - 2017 
pakistan.head()
```

Next, we will plot a line plot by appending .plot() to the pakistan dataframe.

```python
pakistan.plot()
```

We can see in the plot how the cereal yield per ha increased from 856 kg per ha to 3000 kg per ha from 1961 to 2017.

pandas automatically populated the x-axis with the index values (years), and the y-axis with the column values (cereal yield). let's change the type of the index values to integer for plotting.

Also, let's label the x and y axis using plt.title(), plt.ylabel(), and plt.xlabel() as follows:

```python
pakistan.index = pakistan.index.map(int) # let's change the index values of Haiti to type integer for plotting
pakistan.plot(kind='line')

plt.title('Cereal yield (kg/ha) for Pakistan')
plt.ylabel('Cereal Yield')
plt.xlabel('Years')

plt.show()
```

We can clearly notice how the floods in 2010 resulted in low yield. Let's annotate this in the plot by using the plt.text() method.

```python
pakistan.plot(kind='line')

plt.title('Cereal Yield (kg/ha) for Pakistan')
plt.ylabel('Cereal Yield')
plt.xlabel('Years')

# annotate the 2010 Floods. 
# syntax: plt.text(x, y, label)
plt.text(2010, 2500, '2010 Floods') # see note below

plt.show() 
```

Quick note on x and y values in plt.text(x, y, label):

 Since the x-axis (years) is type 'integer', we specified x as a year. The y axis (cereal yield) is type 'integer', so we can just specify the value y = 2500.
    plt.text(2010, 2500, '2010 Floods') # years stored as type int
If the years were stored as type 'string', we would need to specify x as the index position of the year.
    plt.text(2010, 2500, '2010 Floods') # years stored as type int

We can easily add more countries to line plot to make meaningful comparisons of cereal yield for differet countries.

Let's compare the number of immigrants from India and Pakistan from 1961 to 2017.

Get the data set for India and Pakistan, and display dataframe.

```python
df_IP = df_agri.loc[['India', 'Pakistan'], years]
df_IP.head()
```

Let's plot the cereal yield for these countries, so that we can view the comparison in one go. 

```python
df_IP.plot(kind='line')
```

That looks awful. But we can make it a little less confusing. 
pandas plots the indices on the x-axis and the columns as individual lines on the y-axis. Since df_IP is a dataframe with the country as the index and years as the columns, we must first transpose the dataframe using transpose() method to swap the row and columns.

```python
df_IP = df_IP.transpose()
df_IP.head()
```

pandas will auomatically graph the two countries on the same graph. Go ahead and plot the new transposed dataframe.

```python
df_IP.index = df_IP.index.map(int) # let's change the index values of df_IP to type integer for plotting
df_IP.plot(kind='line')

plt.title('Cereal Yield (kg/ha) for India and Pakistan')
plt.ylabel('Cereal Yield')
plt.xlabel('Years')
plt.show() 
```

That looks much better. We can observe year by year comparison of the cereal yeild of the two coutries.

let's compare three countries now by including Bangladesh in the above comparison

```python
df_BIP = df_agri.loc[['Bangladesh','India', 'Pakistan'], years]
df_BIP.head()
```

```python
df_BIP = df_BIP.transpose()
df_BIP.head()
```

```python
df_BIP.index = df_BIP.index.map(int) # let's change the index values of df_BIP to type integer for plotting
df_BIP.plot(kind='line')

plt.title('Cereal Yield (kg/ha) for Bangladesh, India and Pakistan')
plt.ylabel('Cereal Yield')
plt.xlabel('Years')
plt.show() 
```

Bangladesh seems to be doing a better job than the two neighbors.
