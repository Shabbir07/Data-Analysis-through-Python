# Data Analysis and Visalization with Python

### Data : Agriculture value added (%) to GDP 
#### Years : 1960 - 2018
#### Data Source : World Bank 

```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Downloading the dataset
df_value = pd.read_csv('addvalue.csv')
print ('Data read into a pandas dataframe!')
```

```python
# let's see how the data look like
df_value.head()
```

```python
# Determining the size of the dataset
df_value.shape
```

```python
# let's have a look at the whole dataset
df_value
```

```python
# for sake of consistency, let's also make all column labels of type string
df_value.columns = list(map(str, df_value.columns))
```

```python
# set the country name as index - useful for quickly looking up countries using .loc method
df_value.set_index('Country Name', inplace=True)
```

```python
# years that we will be using in this lesson - useful for plotting later on
years = list(map(str, range(1960, 2018)))
print('data dimensions:', df_value.shape)
```

```python
# Let's rename the column 'Country Name' with 'Country' to make it more convenient to approach
df_value.rename(columns={'Country Name':'Country'}, inplace=True)
```

let's view a quick summary of each column in our dataframe using the describe() method.

```python
df_value.describe()
```

The percent value added to GDP by agriculture reflects the development status of a country. A higher percent means that the country's economy is heavily reliant on agriculture and the overall economic status of the country is relatively lower than the countries with a low percent share of agriculture value added to GDP. 

We can see in the above description that the overall average value for 45 countries in this study went down from 35.8 to 14.6 percent from 1960 to 2018. It shows a significant improvement in other sectors of the economy. However, this improvement is mainly due to an even bigger improvement in some of the countries. 
We can compare the changes in agriculture value added through different visualization techniques. It will help us determine which countries are still relaint on agriculture for their economic growth. 

## Visualization

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__)
```

### Line Plots

As we have already set the 'Country' column as the index using set_index() method, therefore we can query a specific country by the country name instead of indexing. 

```python
malaysia = df_value.loc['Malaysia', years] 
malaysia.head()
```

```python
malaysia.plot()
```

It can be seen in the graph that the GDP value added for agriculture went down from 45% in 1961 to less than 8% in 2018. It shows that Malaysian economy has become less reliant on agriculture over the years and more on the other sectors. 
Let's plot data for another country.

```python
somalia = df_value.loc['Somalia', years]
somalia.head()
```

```python
somalia.plot()
```

The case of Somalia is unusual. Agriculture value added was 64% in 1961 and 62% in 2018 so it has remained almost the same but we can see downward trend during 1970s. Its not because the economy switched to other sectors but it was due to severe economic distress. The GDP growth rate in the country was -18% in 1974 (Source: World Bank), hence the sharp dip in the graph for that year. 

```python
# let's plot the progress for Turkey
turkey = df_value.loc['Turkey', years]
turkey.plot()
```

```python
# Let's compare Malaysis and Turkey
df_MT = df_value.loc[['Malaysia', 'Turkey'], years]
df_MT = df_MT.transpose()
df_MT.head()
```

```python
df_MT.index = df_MT.index.map(int) # let's change the index values of df_MT to type integer for plotting
df_MT.plot(kind='line')
```

```python
#Let's add title and label the axes
df_MT.index = df_MT.index.map(int)
df_MT.plot(kind='line')

plt.title('Agriculture Value Added')
plt.ylabel('Percent Value Added')
plt.xlabel('Years')

plt.show()
```

### Histogram

```python
#Plotting the same data on a histogram
df_MT.index = df_MT.index.map(int)
df_MT.plot(kind='hist', figsize=(8,5))

plt.title('Agriculture Value Added')
plt.ylabel('Percent Value Added')
plt.xlabel('Years')

plt.show()
```

### Bar Chart

```python
# Let's compare Pakistan and Bostwana since both had 43% agriculture value added to GDP in 1960. 
df_MP = df_value.loc[['Malaysia', 'Pakistan'], years]
df_MP = df_MP.transpose()
df_MP.head()
```

```python
#plotting the data
df_MP.plot(kind='bar', figsize=(12, 8))

plt.xlabel('Year') 
plt.ylabel('Agriculture Value Added') 
plt.title('Malaysia vs Pakistan') 

plt.show()
```

Let's create a horizantal bar plot

```python
df_MP.plot(kind='barh', figsize=(14, 10))

plt.xlabel('Year') 
plt.ylabel('Agriculture Value Added') 
plt.title('Malaysia vs Pakistan') 

plt.show()
```

It is evident that Malaysia has reduce its reliance on agriculture siginificantly in comparison to agriculture. 

We can compare the data between Gabon and Sri Lanka since both had 32% agriculture value added to GDP in 1960.

```python
df_GS = df_value.loc[['Gabon', 'Sri Lanka'], years]
df_GS = df_GS.transpose()
df_GS.head()
```

```python
#plotting the data
df_GS.plot(kind='barh', figsize=(12, 8))

plt.xlabel('Year') 
plt.ylabel('Agriculture Value Added') 
plt.title('Gabon vs Sri Lanka') 

plt.show()
```
